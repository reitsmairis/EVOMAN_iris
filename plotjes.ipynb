```python
"""
Created on Fri Sep 17 2021

@author: Iris
make figures
"""
import csv
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Retrieve information from NEAT csv

total_fitness_data = []
with open('testing_data_NEAT.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f, delimiter=',', quoting=csv.QUOTE_NONNUMERIC)
    for row in reader:
        total_fitness_data.append(row)
        
enemy, generations = total_fitness_data[0]
total_fitness_data = total_fitness_data[1:]

enemy = int(enemy)
generations = int(generations)
population_size = len(total_fitness_data)
x = range(generations)
total_fitness_data = np.array(total_fitness_data)
best_fitness = total_fitness_data[:,0]
avg_fitness = total_fitness_data[:,1]
stdev_fitness = total_fitness_data[:,2]
```

```python
fig, ax = plt.subplots()
ax.plot(x, best_fitness)
ax.plot(x, avg_fitness)
ax.fill_between(x, 
                avg_fitness - stdev_fitness, 
                avg_fitness + stdev_fitness, alpha=0.2)
ax.set_xlim(0, generations-1)
ax.set_xticks(range(generations))
plt.xlabel('Generation', fontsize=18)
plt.ylabel('Fitness', fontsize=18)
plt.ylim(0,100)
plt.text(0.7*generations, 10, f'Max Fit  = {np.max(total_fitness_data[:,0]).round(2)}')
plt.title(f'Fitness over time against enemy:{int(enemy)}', fontsize=18)
plt.show()
```
